```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import solve
import pendulum

def compute_cost(theta, omega, u):
    cost = np.sum(10 * (theta - np.pi)**2 + 0.1 * omega**2 + 0.1 * u**2)
    return cost

def compute_gradient(theta, omega, u):
    N = len(u) - 1  # Number of time steps
    grad = np.zeros(3 * (N + 1))
    
    for n in range(N + 1):
        idx_theta = n
        idx_omega = (N + 1) + n
        idx_u = 2 * (N + 1) + n
        
        grad[idx_theta] = 20 * (theta[n] - np.pi)
        grad[idx_omega] = 0.2 * omega[n]
        grad[idx_u] = 0.2 * u[n]
    
    return grad

def compute_hessian(N):
    size = 3 * (N + 1)
    H = np.zeros((size, size))
    
    for n in range(N + 1):
        idx_theta = n
        idx_omega = (N + 1) + n
        idx_u = 2 * (N + 1) + n
        
        H[idx_theta, idx_theta] = 20
        H[idx_omega, idx_omega] = 0.2
        H[idx_u, idx_u] = 0.2
    
    return H

# def compute_constraints(theta, omega, u, dt, g_const):
#     N = len(u) - 1  # Number of time steps (N = 300)
#     num_constraints = 2 * N
#     size = 3 * (N + 1)
    
#     G = np.zeros((num_constraints, size))
#     c_vec = np.zeros(num_constraints)
    
#     for n in range(N):
#         # Variable indices (for accessing theta, omega, u)
#         var_idx_n = n
#         var_idx_np1 = n + 1
        
#         # Indices in x_guess (for constructing G)
#         idx_theta_n = n
#         idx_theta_np1 = n + 1
#         idx_omega_n = (N + 1) + n
#         idx_omega_np1 = (N + 1) + n + 1
#         idx_u_n = 2 * (N + 1) + n
        
#         # Constraint c_n^\theta
#         c_theta = theta[var_idx_np1] - theta[var_idx_n] - dt * omega[var_idx_n]
#         c_vec[2 * n] = c_theta
        
#         G[2 * n, idx_theta_n] = -1
#         G[2 * n, idx_theta_np1] = 1
#         G[2 * n, idx_omega_n] = -dt
        
#         # Constraint c_n^\omega
#         sin_theta_n = np.sin(theta[var_idx_n])
#         cos_theta_n = np.cos(theta[var_idx_n])
#         c_omega = omega[var_idx_np1] - omega[var_idx_n] - dt * (u[var_idx_n] - g_const * sin_theta_n)
#         c_vec[2 * n + 1] = c_omega
        
#         G[2 * n + 1, idx_omega_n] = -1
#         G[2 * n + 1, idx_omega_np1] = 1
#         G[2 * n + 1, idx_u_n] = -dt
#         G[2 * n + 1, idx_theta_n] = dt * g_const * cos_theta_n  # Note the positive sign due to derivative
    
#     return G, -c_vec  # Negative c_vec because we want G * Delta x = -c_vec

# def compute_constraints(theta, omega, u, dt, g_const):
#     """
#     Compute the linearized constraints and their Jacobian.

#     Parameters:
#     - theta: Array of angles (size N+1)
#     - omega: Array of angular velocities (size N+1)
#     - u: Array of control inputs (size N+1)
#     - dt: Time step size
#     - g_const: Gravitational constant

#     Returns:
#     - G: Jacobian matrix of constraints (size 2N x 3*(N+1))
#     - c_vec: Constraint residuals vector (size 2N)
#     """
#     N = len(u) - 1  # Ensure N is len(u) - 1, i.e., 300
#     num_constraints = 2 * N
#     size = 3 * (N + 1)

#     G = np.zeros((num_constraints, size))
#     c_vec = np.zeros(num_constraints)

#     for n in range(N):  # n = 0 to N-1 (0 to 299)
#         var_idx_n = n
#         var_idx_np1 = n + 1

#         idx_theta_n = var_idx_n
#         idx_theta_np1 = var_idx_np1
#         idx_omega_n = (N + 1) + var_idx_n
#         idx_omega_np1 = (N + 1) + var_idx_np1
#         idx_u_n = 2 * (N + 1) + var_idx_n

#         # Assertions to prevent out-of-bounds access
#         assert idx_theta_np1 < len(theta), f"theta index {idx_theta_np1} out of bounds for size {len(theta)}"
#         assert idx_theta_n < len(theta), f"theta index {idx_theta_n} out of bounds for size {len(theta)}"
#         assert idx_omega_n < len(omega), f"omega index {idx_omega_n} out of bounds for size {len(omega)}"
#         assert idx_omega_np1 < len(omega), f"omega index {idx_omega_np1} out of bounds for size {len(omega)}"
#         assert idx_u_n < len(u) + 2*(N +1), f"u index {idx_u_n} out of bounds for size {len(u)}"

#         # Constraint c_n^theta: theta_{n+1} - theta_n - dt * omega_n = 0
#         c_theta = theta[idx_theta_np1] - theta[idx_theta_n] - dt * omega[idx_omega_n]
#         c_vec[2 * n] = c_theta

#         G[2 * n, idx_theta_n] = -1
#         G[2 * n, idx_theta_np1] = 1
#         G[2 * n, idx_omega_n] = -dt

#         # Constraint c_n^omega: omega_{n+1} - omega_n - dt * (u_n - g * sin(theta_n)) = 0
#         sin_theta_n = np.sin(theta[idx_theta_n])
#         cos_theta_n = np.cos(theta[idx_theta_n])
#         c_omega = omega[idx_omega_np1] - omega[idx_omega_n] - dt * (u[idx_u_n] - g_const * sin_theta_n)
#         c_vec[2 * n + 1] = c_omega

#         G[2 * n + 1, idx_omega_n] = -1
#         G[2 * n + 1, idx_omega_np1] = 1
#         G[2 * n + 1, idx_u_n] = -dt
#         G[2 * n + 1, idx_theta_n] = dt * g_const * cos_theta_n  # Partial derivative w.r.t theta_n

#     return G, -c_vec  # Negative c_vec because we want G * Delta x = -c_vec


def compute_constraints(theta, omega, u, dt, g_const):
    N = len(u) - 1  # Number of time steps (N = 300)
    num_constraints = 2 * N
    size = 3 * (N + 1)
    
    G = np.zeros((num_constraints, size))
    c_vec = np.zeros(num_constraints)
    
    for n in range(N):
        # Variable indices (for accessing theta, omega, u)
        var_idx_n = n
        var_idx_np1 = n + 1
        
        # Indices in x_guess (for constructing G)
        idx_theta_n = n
        idx_theta_np1 = n + 1
        idx_omega_n = (N + 1) + n
        idx_omega_np1 = (N + 1) + n + 1
        idx_u_n = 2 * (N + 1) + n
        
        # Constraint c_n^\theta
        c_theta = theta[var_idx_np1] - theta[var_idx_n] - dt * omega[var_idx_n]
        c_vec[2 * n] = c_theta
        
        G[2 * n, idx_theta_n] = -1
        G[2 * n, idx_theta_np1] = 1
        G[2 * n, idx_omega_n] = -dt
        
        # Constraint c_n^\omega
        sin_theta_n = np.sin(theta[var_idx_n])
        cos_theta_n = np.cos(theta[var_idx_n])
        c_omega = omega[var_idx_np1] - omega[var_idx_n] - dt * (u[var_idx_n] - g_const * sin_theta_n)
        c_vec[2 * n + 1] = c_omega
        
        G[2 * n + 1, idx_omega_n] = -1
        G[2 * n + 1, idx_omega_np1] = 1
        G[2 * n + 1, idx_u_n] = -dt
        G[2 * n + 1, idx_theta_n] = dt * g_const * cos_theta_n  # Note the positive sign due to derivative
    
    return G, -c_vec  # Negative c_vec because we want G * Delta x = -c_vec

def solve_kkt(H, grad_J, G, c_vec):
    """
    Solve the KKT system to obtain search directions.
    
    Parameters:
    - H: Hessian matrix (size 3*(N+1) x 3*(N+1))
    - grad_J: Gradient vector (size 3*(N+1))
    - G: Jacobian matrix of constraints (size 2N x 3*(N+1))
    - c_vec: Constraint residuals vector (size 2N)
    
    Returns:
    - delta_x: Search direction for variables (size 3*(N+1))
    - delta_lambda: Search direction for Lagrange multipliers (size 2N)
    """
    num_vars = H.shape[0]          # 903
    num_constraints = G.shape[0]  # 600
    
    # Initialize KKT matrix of size (1503, 1503)
    KKT_matrix = np.zeros((num_vars + num_constraints, num_vars + num_constraints))
    
    # Assign H to the top-left block
    KKT_matrix[:num_vars, :num_vars] = H
    
    # Correct indexing: Assign G.T to the top-right block (903, 600)
    KKT_matrix[:num_vars, num_vars:num_vars + num_constraints] = G.T
    
    # Assign G to the bottom-left block (600, 903)
    KKT_matrix[num_vars:num_vars + num_constraints, :num_vars] = G
    
    # The bottom-right block remains zeros
    
    # Assemble the right-hand side
    rhs = -np.concatenate([grad_J, c_vec])
    
    # Solve the KKT system
    try:
        solution = solve(KKT_matrix, rhs)
    except np.linalg.LinAlgError:
        print("KKT matrix is singular. Adding regularization.")
        epsilon = 1e-6
        KKT_matrix[:num_vars, :num_vars] += epsilon * np.eye(num_vars)
        solution = solve(KKT_matrix, rhs)
    
    # Extract delta_x and delta_lambda
    delta_x = solution[:num_vars]
    delta_lambda = solution[num_vars:]
    
    return delta_x, delta_lambda

def compute_constraint_violation(c_vec):
    violation = np.sum(np.abs(c_vec))
    return violation

def filter_line_search(x_guess, p_x, p_lambda, f_current, c_current, compute_cost, compute_constraints, dt, g_const):
    rho = 0.5
    alpha = 1.0
    max_iter = 20
    lambda_guess = np.zeros_like(p_lambda)
    
    for _ in range(max_iter):
        x_trial = x_guess + alpha * p_x
        N = len(x_guess) // 3 - 1
        theta_trial = x_trial[:N + 1]
        omega_trial = x_trial[N + 1:2 * (N + 1)]
        u_trial = x_trial[2 * (N + 1):]
        
        f_trial = compute_cost(theta_trial, omega_trial, u_trial)
        G_trial, c_vec_trial = compute_constraints(theta_trial, omega_trial, u_trial, dt, g_const)
        c_trial_violation = compute_constraint_violation(-c_vec_trial)
        
        if f_trial < f_current or c_trial_violation < c_current:
            return x_trial, p_lambda, f_trial, c_trial_violation, alpha
        
        alpha *= rho
    print("Line search failed.")
    return x_guess, lambda_guess, f_current, c_current, 0.0

def check_convergence(grad_L, c_vec, tol=1e-4):
    optimality = np.linalg.norm(grad_L, np.inf)
    feasibility = np.linalg.norm(c_vec, np.inf)
    return optimality < tol and feasibility < tol

def split_variables(x_guess):
    N = (len(x_guess) // 3) - 1
    theta = x_guess[:N + 1]
    omega = x_guess[N + 1:2 * (N + 1)]
    u = x_guess[2 * (N + 1):]
    # print(theta.shape, omega.shape, u.shape)
    return theta, omega, u

def sqp_solver():
    N = 300
    dt = 0.01
    g_const = 9.81
    max_iter = 100
    tol = 1e-4

    theta = np.zeros(N + 1)
    omega = np.zeros(N + 1)
    u = np.zeros(N + 1)
    
    x_guess = np.concatenate([theta, omega, u])
    lambda_guess = np.zeros(2 * N)
    
    theta, omega, u = split_variables(x_guess)
    f_current = compute_cost(theta, omega, u)
    G, c_vec = compute_constraints(theta, omega, u, dt, g_const)
    c_current = compute_constraint_violation(-c_vec)

    cost_history = [f_current]
    constraint_violation_history = [c_current]
    alpha_history = [1.0]

    for iteration in range(max_iter):
        grad_J = compute_gradient(theta, omega, u)
        H = compute_hessian(N)
        G, c_vec = compute_constraints(theta, omega, u, dt, g_const)
        grad_L = grad_J + G.T @ lambda_guess

        if check_convergence(grad_L, c_vec, tol):
            print(f"Converged in {iteration} iterations.")
            break
        
        delta_x, delta_lambda = solve_kkt(H, grad_J, G, c_vec)
        x_guess, lambda_guess, f_current, c_current, alpha = filter_line_search(
            x_guess, delta_x, delta_lambda, f_current, c_current, compute_cost, compute_constraints, dt, g_const)

        theta, omega, u = split_variables(x_guess)
        cost_history.append(f_current)
        constraint_violation_history.append(c_current)
        alpha_history.append(alpha)
    
    return theta, omega, u, cost_history, constraint_violation_history, alpha_history

def plot_solution(theta, omega, u, dt):
    N = len(u) - 1
    time = np.linspace(0, N * dt, N + 1)

    plt.figure(figsize=(14, 10))
    plt.subplot(3, 1, 1)
    plt.plot(time, theta, label='Angle θ (rad)')
    plt.axhline(np.pi, color='r', linestyle='--', label='Desired θ = π rad')
    plt.title('Pendulum Angle Over Time')
    plt.xlabel('Time [s]')
    plt.ylabel('Angle θ (rad)')
    plt.legend()
    plt.grid(True)

    plt.subplot(3, 1, 2)
    plt.plot(time, omega, label='Angular Velocity ω (rad/s)')
    plt.title('Pendulum Angular Velocity Over Time')
    plt.xlabel('Time [s]')
    plt.ylabel('Angular Velocity ω (rad/s)')
    plt.legend()
    plt.grid(True)

    plt.subplot(3, 1, 3)
    plt.plot(time, u, label='Control Input u (N·m)')
    plt.title('Control Input Over Time')
    plt.xlabel('Time [s]')
    plt.ylabel('Control Input u (N·m)')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

def plot_constraint_violation(constraint_violation_hist):
    plt.figure(figsize=(8, 6))
    plt.plot(range(len(constraint_violation_hist)), constraint_violation_hist, marker='o', color='orange')
    plt.title('Constraint Violation per Iteration')
    plt.xlabel('Iteration')
    plt.ylabel('Total Constraint Violation')
    plt.yscale('log')
    plt.grid(True)
    plt.show()

def plot_cost_history(cost_hist):
    plt.figure(figsize=(8, 6))
    plt.plot(range(len(cost_hist)), cost_hist, marker='o', color='green')
    plt.title('Cost Function Value per Iteration')
    plt.xlabel('Iteration')
    plt.ylabel('Cost J')
    plt.yscale('log')
    plt.grid(True)
    plt.show()

def plot_alpha_history(alpha_hist):
    plt.figure(figsize=(8, 6))
    plt.plot(range(len(alpha_hist)), alpha_hist, marker='o', color='purple')
    plt.title('Step Size α per Iteration')
    plt.xlabel('Iteration')
    plt.ylabel('Step Size α')
    plt.grid(True)
    plt.show()

# Execute the SQP solver
theta_opt, omega_opt, u_opt, cost_hist, constraint_violation_hist, alpha_hist = sqp_solver()

# Plot the solution
plot_solution(theta_opt, omega_opt, u_opt, dt=0.01)

# Plot the constraint violation history
plot_constraint_violation(constraint_violation_hist)

# Plot the cost history
plot_cost_history(cost_hist)

# Plot the alpha history
plot_alpha_history(alpha_hist)

# Animate the pendulum motion using the optimized control
x0 = np.array([[0], [0]])
u_for_animation = u_opt[:-1].reshape(1, -1)
pendulum.animate_robot(x0, u_for_animation)
```
