```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import solve
import pendulum

def compute_cost(theta, omega, u):
    """
    Compute the cost function J.

    Parameters:
    - theta: Array of angles (size N+1)
    - omega: Array of angular velocities (size N+1)
    - u: Array of control inputs (size N+1)

    Returns:
    - cost: Scalar value of the cost function
    """
    cost = np.sum(10 * (theta - np.pi)**2 + 0.1 * omega**2 + 0.1 * u**2)
    return cost

def compute_gradient(theta, omega, u):
    """
    Compute the gradient of the cost function with respect to x.

    Parameters:
    - theta: Array of angles (size N+1)
    - omega: Array of angular velocities (size N+1)
    - u: Array of control inputs (size N+1)

    Returns:
    - grad: Gradient vector (size 3*(N+1))
    """
    N_plus_1 = len(theta)
    grad = np.zeros(3 * N_plus_1)
    
    # Compute gradients for theta
    grad[0::3] = 20 * (theta - np.pi)
    
    # Compute gradients for omega
    grad[1::3] = 0.2 * omega
    
    # Compute gradients for u
    grad[2::3] = 0.2 * u
    
    return grad

from scipy.linalg import block_diag
def compute_hessian(N):
    # Define the 3x3 block for theta, omega, and u variables
    block = np.array([
        [20,   0,   0],
        [ 0, 0.2,   0],
        [ 0,   0, 0.2]
    ])
    
    # Create a list of these blocks for each of the N+1 time steps
    blocks = [block] * (N + 1)
    
    # Construct the block diagonal Hessian matrix
    H = block_diag(*blocks)
    
    return H

def compute_constraints(theta, omega, u, dt, g_const):
    """
    Compute the dynamic constraints and their Jacobian matrix G based on interleaved variable ordering,
    including initial conditions.

    Parameters:
    - theta: Array of angles (size N+1)
    - omega: Array of angular velocities (size N+1)
    - u: Array of control inputs (size N+1)
    - dt: Time step size
    - g_const: Gravitational constant

    Returns:
    - G: Jacobian matrix of constraints (size 2N + 2 x 3*(N+1))
    - c_vec: Constraint residuals vector (size 2N + 2)
    """
    N = len(u) - 1  # Number of time steps
    num_constraints = 2 * N + 2  # Added 2 for initial conditions
    size = 3 * (N + 1)

    G = np.zeros((num_constraints, size))
    c_vec = np.zeros(num_constraints)

    for n in range(N):
        # Current and next indices for theta, omega, and u
        idx_theta_n = 3 * n
        idx_omega_n = 3 * n + 1
        idx_u_n = 3 * n + 2

        idx_theta_np1 = 3 * (n + 1)
        idx_omega_np1 = 3 * (n + 1) + 1
        # u_np1 = 3 * (n + 1) + 2  # Not used in constraints

        # Constraint 1: theta_{n+1} - theta_n - dt * omega_n = 0
        c_theta = theta[n + 1] - theta[n] - dt * omega[n]
        c_vec[2 * n] = c_theta

        # Populate Jacobian for c_theta
        G[2 * n, idx_theta_n] = -1
        G[2 * n, idx_theta_np1] = 1
        G[2 * n, idx_omega_n] = -dt

        # Constraint 2: omega_{n+1} - omega_n - dt * (u_n - g * sin(theta_n)) = 0
        c_omega = omega[n + 1] - omega[n] - dt * (u[n] - g_const * np.sin(theta[n]))
        c_vec[2 * n + 1] = c_omega

        # Populate Jacobian for c_omega
        G[2 * n + 1, idx_theta_n] = dt * g_const * np.cos(theta[n])
        G[2 * n + 1, idx_omega_n] = -1
        G[2 * n + 1, idx_omega_np1] = 1
        G[2 * n + 1, idx_u_n] = -dt

    # Add initial condition constraints
    # Constraint 3: theta_0 = 0
    G[-2, 0] = 1  # Coefficient for theta_0
    c_vec[-2] = theta[0]  # Should be zero

    # Constraint 4: omega_0 = 0
    G[-1, 1] = 1  # Coefficient for omega_0
    c_vec[-1] = omega[0]  # Should be zero

    print("Jacobian Matrix G:\n", G)
    print("Constraint Vector c_vec:\n", c_vec)
    return G, c_vec  # Negative c_vec for KKT formulation


def solve_kkt(H, grad_J, G, c_vec):
    """
    Solve the Karush-Kuhn-Tucker (KKT) system to obtain search directions.

    Parameters:
    - H: Hessian matrix (size 3*(N+1) x 3*(N+1))
    - grad_J: Gradient vector (size 3*(N+1))
    - G: Jacobian matrix of constraints (size 2N x 3*(N+1))
    - c_vec: Constraint residuals vector (size 2N)

    Returns:
    - delta_x: Search direction for variables (size 3*(N+1))
    - delta_lambda: Search direction for Lagrange multipliers (size 2N)
    """
    num_vars = H.shape[0]          # 3*(N+1)
    num_constraints = G.shape[0]  # 2N
    
    # Initialize KKT matrix of size (num_vars + num_constraints) x (num_vars + num_constraints)
    KKT_matrix = np.zeros((num_vars + num_constraints, num_vars + num_constraints))
    
    # Assign H to the top-left block
    KKT_matrix[:num_vars, :num_vars] = H
    
    # Assign G.T to the top-right block
    KKT_matrix[:num_vars, num_vars:num_vars + num_constraints] = G.T
    
    # Assign G to the bottom-left block
    KKT_matrix[num_vars:num_vars + num_constraints, :num_vars] = G
    
    # The bottom-right block remains zero
    
    # Assemble the right-hand side
    rhs = - np.concatenate([grad_J, c_vec])
    
    # Solve the KKT system
    try:
        solution = solve(KKT_matrix, rhs)
    except np.linalg.LinAlgError:
        # Regularization for singular matrices
        epsilon = 1e-6
        KKT_matrix[:num_vars, :num_vars] += epsilon * np.eye(num_vars)
        solution = solve(KKT_matrix, rhs)
    
    # Extract delta_x and delta_lambda
    delta_x = solution[:num_vars]
    delta_lambda = solution[num_vars:]
    
    return delta_x, delta_lambda

def compute_constraint_violation(c_vec):
    """
    Compute the total constraint violation.

    Parameters:
    - c_vec: Constraint residuals vector

    Returns:
    - violation: Scalar representing the total constraint violation
    """
    violation = np.sum(np.abs(c_vec))
    return violation

def simple_filter_line_search(x_guess, delta_x, lambda_guess, delta_lambda, f_best, g_best, compute_cost, compute_constraints, dt, g_const, rho=0.5, alpha_init=1.0, max_iter=10):
    """
    Perform the simple filter line search to find an appropriate step size alpha.
    
    Parameters:
    - x_guess: Current decision variables (numpy array)
    - delta_x: Search direction (numpy array)
    - f_best: Current best cost (float)
    - g_best: Current best constraint violation (float)
    - compute_cost: Function to compute cost
    - compute_constraints: Function to compute constraints and G
    - dt: Time step size
    - g_const: Gravitational constant
    - rho: Step size reduction factor (default 0.5)
    - alpha_init: Initial step size (default 1.0)
    - max_iter: Maximum number of iterations (default 20)
    
    Returns:
    - x_trial: Updated decision variables after line search
    - lambda_new: Updated Lagrange multipliers (placeholder, needs proper handling)
    - f_trial: Cost at x_trial
    - c_trial_violation: Constraint violation at x_trial
    - alpha: Step size used
    """
    alpha = alpha_init
    f_best_current = f_best
    g_best_current = g_best
    
    for _ in range(max_iter):
        x_trial = x_guess + alpha * delta_x
        theta_trial, omega_trial, u_trial = split_variables(x_trial)
        f_trial = compute_cost(theta_trial, omega_trial, u_trial)
        __, c_vec_trial = compute_constraints(theta_trial, omega_trial, u_trial, dt, g_const)
        g_trial_violation = compute_constraint_violation(c_vec_trial)
        
        if f_trial < f_best_current or g_trial_violation < g_best_current:
            # update the best
            f_best_new = f_trial if f_trial < f_best_current else f_best_current
            g_best_new = g_trial_violation if g_trial_violation < g_best_current else g_best_current
            
            # update lambda
            lambda_new = (1 - alpha) * lambda_guess + alpha * delta_lambda
            return x_trial, lambda_new, f_best_new, g_best_new, alpha
        
        alpha *= rho
    
    print("No acceptable alpha found.", alpha)
    # If no acceptable alpha found, return the last trial
    return x_trial, lambda_guess, f_trial, g_trial_violation, alpha


def check_convergence(grad_L, c_vec, tol=1e-4):
    """
    Check if the optimization has converged based on gradient and constraint violation.

    Parameters:
    - grad_L: Gradient of the Lagrangian (numpy array)
    - c_vec: Constraint residuals vector (numpy array)
    - tol: Tolerance for convergence

    Returns:
    - converged: Boolean indicating convergence
    """
    optimality = np.linalg.norm(grad_L, np.inf)
    feasibility = np.linalg.norm(c_vec, np.inf)
    return optimality < tol and feasibility < tol

def split_variables(x_guess):
    """
    Split the decision variable vector into theta, omega, and u.

    Parameters:
    - x_guess: Decision variables vector (numpy array)

    Returns:
    - theta: Array of angles (size N+1)
    - omega: Array of angular velocities (size N+1)
    - u: Array of control inputs (size N+1)
    """
    theta = x_guess[0::3]
    omega = x_guess[1::3]
    u = x_guess[2::3]
    return theta, omega, u


def sqp_solver():
    """
    Perform Sequential Quadratic Programming (SQP) to optimize the pendulum control.
    
    Returns:
    - theta_opt: Optimized angles (numpy array)
    - omega_opt: Optimized angular velocities (numpy array)
    - u_opt: Optimized control inputs (numpy array)
    - cost_history: List of cost values per iteration
    - constraint_violation_history: List of constraint violations per iteration
    - alpha_history: List of step sizes per iteration
    """
    # Parameters
    N = 300
    dt = 0.01
    g_const = 9.81
    max_iter = 100
    tol = 1e-4

    # Initialize variables
    x_guess = np.zeros(3 * (N + 1))
    # Initialize Lagrange multipliers (not used in line search but line search updates it)
    lambda_guess = np.zeros(2 * N + 2)
    
    f_best = np.inf
    g_best = np.inf
    
    # Compute initial cost and constraints
    theta, omega, u = split_variables(x_guess)
    f_current = compute_cost(theta, omega, u)
    # G is the Jacobian matrix; g is the vector g(x)
    G, g = compute_constraints(theta, omega, u, dt, g_const)
    g_current_violation = compute_constraint_violation(g)
    
    # Update f_best and c_best with initial values
    f_best = f_current if f_current < f_best else f_best
    g_best = g_current_violation if g_current_violation < g_best else g_best
    
    # Histories for plotting
    cost_history = [f_current]
    constraint_violation_history = [g_current_violation]
    alpha_history = [1.0]
    
    # Optimization loop
    for iteration in range(1, max_iter + 1):
        # Compute gradient and Hessian
        grad_J = compute_gradient(theta, omega, u)
        H = compute_hessian(N)
        
        # Compute constraints and Jacobian
        G, g = compute_constraints(theta, omega, u, dt, g_const)
        
        # Compute gradient of Lagrangian (assuming lambda_guess = 0 initially)
        grad_L = grad_J + G.T @ lambda_guess
        
        # Check convergence
        if check_convergence(grad_L, g, tol):
            print(f"Converged in {iteration - 1} iterations.")
            break
        
        # Solve KKT system to get search directions
        delta_x, delta_lambda = solve_kkt(H, grad_J, G, g)
        
        # Perform the simple filter line search to find step size alpha
        # which, takes the direction delta_x and delta_lambda, and returns
        # the updated f and g values
        x_trial, lambda_new, f_trial, g_trial_violation, alpha = simple_filter_line_search(
            x_guess, delta_x, lambda_guess, delta_lambda, f_best, g_best, compute_cost, compute_constraints, dt, g_const
        )
        
        # Update f_best and c_best if necessary
        if f_trial < f_best:
            f_best = f_trial
        if g_trial_violation < g_best:
            g_best = g_trial_violation
        
        # Update variables
        x_guess = x_trial
        lambda_guess = lambda_new  # Update Lagrange multipliers appropriately
        
        # Split variables for next iteration
        theta, omega, u = split_variables(x_guess)
        
        # Update histories
        cost_history.append(f_trial)
        constraint_violation_history.append(g_trial_violation)
        alpha_history.append(alpha)
        
        # Print iteration details
        print(f"Iteration {iteration}: Cost = {f_trial:.6f}, Constraint Violation = {g_trial_violation:.6f}, Alpha = {alpha:.6f}")
    
    else:
        print(f"Reached maximum iterations ({max_iter}) without full convergence.")
    
    return theta, omega, u, cost_history, constraint_violation_history, alpha_history


def plot_solution(theta, omega, u, dt):
    """
    Plot the optimized angles, angular velocities, and control inputs over time.

    Parameters:
    - theta: Optimized angles (numpy array)
    - omega: Optimized angular velocities (numpy array)
    - u: Optimized control inputs (numpy array)
    - dt: Time step size
    """
    N = len(u) - 1
    time = np.linspace(0, N * dt, N + 1)

    plt.figure(figsize=(14, 10))
    
    # Plot Theta
    plt.subplot(3, 1, 1)
    plt.plot(time, theta, label='Angle θ (rad)', color='blue')
    plt.axhline(np.pi, color='red', linestyle='--', label='Desired θ = π rad')
    plt.title('Pendulum Angle Over Time')
    plt.xlabel('Time [s]')
    plt.ylabel('Angle θ (rad)')
    plt.legend()
    plt.grid(True)
    
    # Plot Omega
    plt.subplot(3, 1, 2)
    plt.plot(time, omega, label='Angular Velocity ω (rad/s)', color='green')
    plt.title('Pendulum Angular Velocity Over Time')
    plt.xlabel('Time [s]')
    plt.ylabel('Angular Velocity ω (rad/s)')
    plt.legend()
    plt.grid(True)
    
    # Plot Control Input
    plt.subplot(3, 1, 3)
    plt.plot(time, u, label='Control Input u (N·m)', color='purple')
    plt.title('Control Input Over Time')
    plt.xlabel('Time [s]')
    plt.ylabel('Control Input u (N·m)')
    plt.legend()
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()

def plot_constraint_violation(constraint_violation_hist):
    """
    Plot the constraint violation history over iterations.

    Parameters:
    - constraint_violation_hist: List of constraint violations per iteration
    """
    plt.figure(figsize=(8, 6))
    plt.plot(range(len(constraint_violation_hist)), constraint_violation_hist, marker='o', color='orange')
    plt.title('Constraint Violation per Iteration')
    plt.xlabel('Iteration')
    plt.ylabel('Total Constraint Violation')
    plt.yscale('log')
    plt.grid(True)
    plt.show()

def plot_cost_history(cost_hist):
    """
    Plot the cost function history over iterations.

    Parameters:
    - cost_hist: List of cost values per iteration
    """
    plt.figure(figsize=(8, 6))
    plt.plot(range(len(cost_hist)), cost_hist, marker='o', color='green')
    plt.title('Cost Function Value per Iteration')
    plt.xlabel('Iteration')
    plt.ylabel('Cost J')
    plt.yscale('log')
    plt.grid(True)
    plt.show()

def plot_alpha_history(alpha_hist):
    """
    Plot the step size alpha history over iterations.

    Parameters:
    - alpha_hist: List of step sizes per iteration
    """
    plt.figure(figsize=(8, 6))
    plt.plot(range(len(alpha_hist)), alpha_hist, marker='o', color='purple')
    plt.title('Step Size α per Iteration')
    plt.xlabel('Iteration')
    plt.ylabel('Step Size α')
    plt.grid(True)
    plt.show()

# ===============================
# 4. Execute the SQP Solver and Visualize Results
# ===============================

def main():
    # Execute the SQP solver
    theta_opt, omega_opt, u_opt, cost_hist, constraint_violation_hist, alpha_hist = sqp_solver()
    # print(theta_opt)
    # Plot the optimized solution
    plot_solution(theta_opt, omega_opt, u_opt, dt=0.01)
    
    # Plot constraint violation history
    plot_constraint_violation(constraint_violation_hist)
    
    # Plot cost history
    plot_cost_history(cost_hist)
    
    # Plot alpha history
    plot_alpha_history(alpha_hist)
    
    # Animate the pendulum motion using the optimized control inputs
    x0 = np.array([[0], [0]])  # Initial state: theta=0 rad, omega=0 rad/s
    
    # Prepare control inputs for animation (exclude the last control input u_N)
    u_for_animation = u_opt[:-1].reshape(1, -1)  # Shape: (1, N)
    # print(u_opt.shape)
    # print(u_for_animation)
    
    # Animate the pendulum motion
    pendulum.animate_robot(x0, u_for_animation)

if __name__ == "__main__":
    main()
```
