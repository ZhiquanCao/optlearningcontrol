Consider the following simplified linear model of the drone
$$\begin{align}
    x_{n+1} = A
    x_n + B u_n
\end{align}$$

where the components of the state $x_n$ correspond to the horizontal position and velocity of the quadrotor, its vertical position and velocity and its orientation and angular velocity. The control vector
$u_n$ contains the forces produced by the rotors (our control inputs).
The module ```quadrotor.py``` defines the matrices $A$ and $B$ and code to generate a simulation of the drone.

We want to generate a control input will move the drone towards the point $(3,3)$ starting from $(0,0)$. In order to
do so, we define a cost that penalizes the distance to the goal at each time step while minimizing velocities and the amount of trust needed,
i.e. we want to solve the following optimal control problem
$$\begin{align}
&\min_{x_0, u_0, x_1, u_1, \cdots} \frac{1}{2}\sum_{n=0}^{N} (x_n - x_{desired})^T Q (x_n - x_{desired}) + u_n^T R u_n\\
\textrm{subject to}\ \ & x_{n+1} = A x_{n} + B u_n\\
& x_0 = [0,0,0,0,0,0]^T
\end{align}$$
where $x_{desired} = [3,0,3,0,0,0]^T$

1. Write down the KKT conditions for the problem
2. Write code to solve the problem for N=500 time steps (you will need to find diagonal matrices $Q>0$ and $R>0$ to create a nice movement) by solving the KKT system of linear equations. Do NOT inverse the KKT matrix, instead use the [NumPy solve function](https://numpy.org/doc/stable/reference/generated/numpy.linalg.solve.html) which is numerically more stable and efficient.
4. Show plots of all the states of the robot as a function of time
5. Show plots of the optimal control as a function of time

````python


# Parameters
N = 500  # Number of time steps
dt = 0.01  # Time step


# Desired state and initial state
x_desired = np.array([3, 0, 3, 0, 0, 0])  # Desired state [x, dx, z, dz, phi, dphi]
x0 = np.zeros(6)  # Initial state

# Define diagonal matrices Q and R
Q = np.diag([10, 0.5, 10, 0.5, 100, 20])  # Penalizes state errors
R = np.diag([2, 1])  # Penalizes control effort

# Initialize variables
x = np.zeros((N + 1, 6))  # States trajectory
u = np.zeros((N, 2))  # Control inputs
x[0] = x0  # Initial state

# Backward pass: calculate P and K matrices according to the riccati equation from the slides
P = np.zeros((N + 1, 6, 6))
K = np.zeros((N, 2, 6))
P[N] = Q  # Terminal cost

# Backward computation of P and K matrices
for n in range(N - 1, -1, -1):
    Pn = P[n + 1]
    K[n] = np.linalg.solve(R + quadrotor.B.T @ Pn @ quadrotor.B, quadrotor.B.T @ Pn @ quadrotor.A)  # Kn = (R + quadrotor.B^T Pn quadrotor.B)^-1 quadrotor.B^T Pn quadrotor.A
    P[n] = Q + quadrotor.A.T @ Pn @ quadrotor.A - quadrotor.A.T @ Pn @ quadrotor.B @ K[n]  # Pn-1 = Q + quadrotor.A^T Pn quadrotor.A - quadrotor.A^T Pn quadrotor.B Kn

# Forward pass: calculate the state and control trajectories
for n in range(N):
    u[n] = -K[n] @ (x[n] - x_desired)  # un = -Kn xn
    x[n + 1] = quadrotor.A @ x[n] + quadrotor.B @ u[n]  # xn+1 = quadrotor.A xn + quadrotor.B un

# Plot results
time = np.linspace(0, N * dt, N + 1)

plt.figure(figsize=(14, 10))

# Plot states
plt.subplot(3, 1, 1)
plt.plot(time, x[:, 0], label='x (position)')
plt.plot(time, x[:, 1], label='dx (velocity)')
plt.plot(time, x[:, 2], label='z (altitude)')
plt.plot(time, x[:, 3], label='dz (vertical speed)')
plt.plot(time, x[:, 4], label='phi (pitch)')
plt.plot(time, x[:, 5], label='dphi (angular speed)')
plt.title('States of the Drone over Time')
plt.xlabel('Time [s]')
plt.ylabel('States')
plt.legend()

# Plot control inputs
plt.subplot(3, 1, 2)
plt.plot(time[:-1], u[:, 0], label='u1 (thrust 1)')
plt.plot(time[:-1], u[:, 1], label='u2 (thrust 2)')
plt.title('Control Inputs over Time')
plt.xlabel('Time [s]')
plt.ylabel('Control Inputs')
plt.legend()

plt.tight_layout()
plt.show()

````

## Code for Hw2

````python
from qpsolvers import solve_qp

xdim = 6
udim = 2
N = 500
nvars = N*udim + (N+1)*xdim # total number of variables

x_desired = np.array([3, 0, 3, 0, 0, 0])  # [x, dx, z, dz, phi, dphi]
x0 = np.zeros(6)  # Initial state

Q = np.diag([10, 0.5, 10, 0.5, 100, 20])
R = np.diag([2, 1])

# Initialize matrices
C = np.zeros(((N + 1) * xdim, nvars))
b = np.zeros((N + 1) * xdim)

# Cx = d
C[0:xdim, 0:xdim] = np.eye(xdim)
b[0:xdim] = x0  # x0 is the initial state

for i in range(N):
    # Indices for variables
    x_start = i * (xdim + udim)
    x_end = x_start + xdim
    u_start = x_end
    u_end = u_start + udim
    x_next_start = (i + 1) * (xdim + udim)
    x_next_end = x_next_start + xdim

    # Rows corresponding to the dynamics constraints
    row_start = xdim * (i + 1)
    row_end = row_start + xdim

    # Build the constraint: A x_n + B u_n - x_{n+1} = 0
    C[row_start:row_end, x_start:x_end] = quadrotor.A
    C[row_start:row_end, u_start:u_end] = quadrotor.B
    C[row_start:row_end, x_next_start:x_next_end] = -np.eye(xdim)

# g: g.T x term in the optimization
q = np.zeros(nvars)
q_vec = -1 * Q @ x_desired  # Precompute - Q x_desired according to the second example from the repo

for i in range(N + 1):
    x_start = i * (xdim + udim)
    x_end = x_start + xdim

    q[x_start:x_end] = q_vec

# The function to get the cost matrix from the example in the repo
def get_cost_matrix_G(Q, R):
    '''
    we are making this a function so we can reuse this later in the next cells when we play with costs
    '''
    G = np.zeros((nvars, nvars))
    for i in range(N):
        G[i*(xdim+udim):i*(xdim+udim)+xdim,i*(xdim+udim):i*(xdim+udim)+xdim] = Q
        G[i*(xdim+udim)+xdim:i*(xdim+udim)+xdim+udim, i*(xdim+udim)+xdim:i*(xdim+udim)+xdim+udim] = R
    G[N*(xdim+udim):N*(xdim+udim)+xdim,N*(xdim+udim):N*(xdim+udim)+xdim] = Q
    return G

P = get_cost_matrix_G(Q,R)

lb = np.full(nvars, -np.inf)
ub = np.full(nvars, np.inf)
lb[1::xdim] = -2
ub[1::xdim] = 2
lb[3::xdim] = -2
ub[3::xdim] = 2
lb[6::xdim] = -5
ub[6::xdim] = 5
lb[7::xdim] = -5
ub[7::xdim] = 5

print(P.shape)
print(q.shape)
print(C.shape)
print(b.shape)
# Solve qp: refer to https://github.com/qpsolvers/qpsolvers
x = solve_qp(P, q, np.zeros(nvars), np.array([np.inf]), C, b, lb, ub, solver="cvxopt")
````

````python
# Initialize arrays to store the results
state_trajectory = np.zeros((N + 1, xdim))
control_trajectory = np.zeros((N, udim))

for i in range(N):
    x_start = i * (xdim + udim)
    u_start = x_start + xdim
    x_end = x_start + xdim

    state_trajectory[i, :] = x[x_start:x_end]
    control_trajectory[i, :] = x[u_start:u_start + udim]

# Last state
x_start = N * (xdim + udim)
state_trajectory[N, :] = x[x_start:x_start + xdim]
````

````python
import matplotlib.pyplot as plt
%matplotlib inline
t = np.linspace(0, N * dt, N + 1)

# Plot positions
plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.plot(t, state_trajectory[:, 0], label='x position')
plt.plot(t, state_trajectory[:, 2], label='z position')
plt.ylabel('Position')
plt.legend()

# Plot velocities
plt.subplot(3, 1, 2)
plt.plot(t, state_trajectory[:, 1], label='x velocity')
plt.plot(t, state_trajectory[:, 3], label='z velocity')
plt.ylabel('Velocity')
plt.legend()

# Plot controls
plt.subplot(3, 1, 3)
plt.step(t[:-1], control_trajectory[:, 0], label='Control u1')
plt.step(t[:-1], control_trajectory[:, 1], label='Control u2')
plt.xlabel('Time [s]')
plt.ylabel('Control Inputs')
plt.legend()

plt.tight_layout()
plt.show()
````

````python

x0 = np.zeros((6,1))
u = control_trajectory
quadrotor.animate_robot(x0, u.transpose(), x_desired)
````
